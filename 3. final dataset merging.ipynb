```python
# Import relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os

# Import garbage collection module for memory management
try:
    import gc
    gc_available = True
except ImportError:
    gc_available = False
    print("Warning: gc module not available, skipping garbage collection")

print("="*60)
print("DATASET LOADING AND CLEANING")
print("="*60)



# First, let's see what columns we have in the cleaned hospital dataset
print("\nAvailable hospital dataset columns:")
for i, col in enumerate(hospital_cleaned.columns):
    print(f"  {i+1:2d}. {col}")

# Define the mapping from original columns to our required features
column_mapping = {
    # Original column names → Required feature names
    'elective_g_a_ordinary_admissions_ffces': 'elective_overnight_admissions',
    'elective_g_a_daycase_admissions_ffces': 'elective_daycase_admissions',
    'total_non_elective_g_a_admissions_ffces': 'emergency_admissions',
    'gp_referrals_made_all_specialties': 'gp_referrals_made_total',
    'gp_referrals_seen_all_specialties': 'gp_referrals_seen_total',
    'all_1st_outpatient_attendances_g_a': 'first_outpatient_attendances',
    'elective_g_a_total_admissions_ffces': 'elective_total_admissions',
    # Basic columns
    'org_code': 'ORG CODE',
    'org_name': 'TRUST',
    'total_activity': 'total_hospital_activity'
}

# Check which columns are available and rename them
print("\nChecking for required columns...")
available_mappings = {}
missing_columns = []

for original_col, new_col in column_mapping.items():
    if original_col in hospital_cleaned.columns:
        available_mappings[original_col] = new_col
        print(f"✓ Found: {original_col} → {new_col}")
    else:
        missing_columns.append(original_col)
        print(f"✗ Missing: {original_col}")

# Rename the available columns
if available_mappings:
    hospital_cleaned = hospital_cleaned.rename(columns=available_mappings)
    print(f"\n✅ Renamed {len(available_mappings)} columns successfully")
else:
    print("\n⚠️ Warning: No target columns found! Proceeding with basic merge only.")

def clean_numeric_column(series, column_name):
    """Clean and convert a single column to numeric"""
    print(f"    Cleaning {column_name}...")

    # Convert to string first
    series_clean = series.astype(str)

    # Replace common non-numeric values
    replacements = {
        'nan': '0', 'NaN': '0', 'NULL': '0', 'null': '0',
        '': '0', '-': '0', 'N/A': '0', 'n/a': '0', 'None': '0'
    }
    for old_val, new_val in replacements.items():
        series_clean = series_clean.replace(old_val, new_val)

    # Remove non-numeric characters (keep digits, decimals, minus signs)
    series_clean = series_clean.str.replace(r'[^\d.-]', '', regex=True)

    # Handle empty strings after cleaning
    series_clean = series_clean.replace('', '0')

    # Convert to numeric
    series_numeric = pd.to_numeric(series_clean, errors='coerce')

    # Fill NaN with 0 and ensure non-negative
    series_numeric = series_numeric.fillna(0).abs()

    non_zero_count = (series_numeric > 0).sum()
    print(f"      → {non_zero_count:,} non-zero values out of {len(series_numeric):,}")

    return series_numeric

def create_required_features(df):
    """Create all required features for ML analysis"""
    print("\n" + "="*50)
    print("CREATING REQUIRED ML FEATURES")
    print("="*50)

    # List of columns that should be numeric
    numeric_columns = [
        'elective_overnight_admissions', 'elective_daycase_admissions',
        'emergency_admissions', 'gp_referrals_made_total', 'gp_referrals_seen_total',
        'first_outpatient_attendances', 'elective_total_admissions', 'total_hospital_activity'
    ]

    # Clean numeric columns
    print("1. Cleaning numeric columns...")
    for col in numeric_columns:
        if col in df.columns:
            df[col] = clean_numeric_column(df[col], col)

    # Create derived features
    print("\n2. Creating derived features...")

    # Feature 1: daycase_to_overnight_ratio
    if 'elective_daycase_admissions' in df.columns and 'elective_overnight_admissions' in df.columns:
        print("   ✓ Creating daycase_to_overnight_ratio...")
        df['daycase_to_overnight_ratio'] = (
            df['elective_daycase_admissions'] / (df['elective_overnight_admissions'] + 1)
        )
        print(f"     → Mean ratio: {df['daycase_to_overnight_ratio'].mean():.2f}")
    else:
        print("   ✗ Cannot create daycase_to_overnight_ratio - missing base columns")
        df['daycase_to_overnight_ratio'] = 0

    # Feature 2: emergency_to_elective_ratio
    if 'emergency_admissions' in df.columns:
        # Calculate total elective if not available
        if 'elective_total_admissions' not in df.columns:
            if 'elective_overnight_admissions' in df.columns and 'elective_daycase_admissions' in df.columns:
                df['elective_total_admissions'] = df['elective_overnight_admissions'] + df['elective_daycase_admissions']
                print("   ✓ Created elective_total_admissions from components")

        if 'elective_total_admissions' in df.columns:
            print("   ✓ Creating emergency_to_elective_ratio...")
            df['emergency_to_elective_ratio'] = (
                df['emergency_admissions'] / (df['elective_total_admissions'] + 1)
            )
            print(f"     → Mean ratio: {df['emergency_to_elective_ratio'].mean():.2f}")
        else:
            print("   ✗ Cannot create emergency_to_elective_ratio - missing elective totals")
            df['emergency_to_elective_ratio'] = 0
    else:
        print("   ✗ Cannot create emergency_to_elective_ratio - missing emergency_admissions")
        df['emergency_to_elective_ratio'] = 0

    # Feature 3: referral_conversion_rate
    if 'gp_referrals_seen_total' in df.columns and 'gp_referrals_made_total' in df.columns:
        print("   ✓ Creating referral_conversion_rate...")
        df['referral_conversion_rate'] = (
            df['gp_referrals_seen_total'] / (df['gp_referrals_made_total'] + 1)
        )
        print(f"     → Mean conversion rate: {df['referral_conversion_rate'].mean():.2f}")
    else:
        print("   ✗ Cannot create referral_conversion_rate - missing referral columns")
        df['referral_conversion_rate'] = 0

    # Feature 4: activity_diversity_index
    activity_base_columns = [
        'elective_overnight_admissions', 'elective_daycase_admissions',
        'emergency_admissions', 'first_outpatient_attendances'
    ]

    available_activity_cols = [col for col in activity_base_columns if col in df.columns]

    if len(available_activity_cols) >= 3:
        print(f"   ✓ Creating activity_diversity_index from {len(available_activity_cols)} activity types...")

        # Get activity data for diversity calculation
        activity_data = df[available_activity_cols].fillna(0)

        # Calculate Shannon diversity index
        diversity_scores = []
        for i in range(len(df)):
            row_activities = activity_data.iloc[i].values
            total_activity = row_activities.sum()

            if total_activity > 0:
                # Calculate proportions
                proportions = row_activities / total_activity
                # Remove zero proportions for log calculation
                proportions = proportions[proportions > 0]
                # Calculate Shannon entropy
                entropy = -(proportions * np.log(proportions)).sum()
                diversity_scores.append(entropy)
            else:
                diversity_scores.append(0)

        df['activity_diversity_index'] = diversity_scores
        print(f"     → Mean diversity index: {df['activity_diversity_index'].mean():.2f}")
    else:
        print(f"   ✗ Cannot create activity_diversity_index - only {len(available_activity_cols)} activity columns available")
        df['activity_diversity_index'] = 0

    print("="*50)
    return df

# Create features in the hospital dataset
hospital_cleaned = create_required_features(hospital_cleaned)

print("="*60)
print("DATASET MERGING")
print("="*60)

# Check for common ORG CODEs and date ranges between datasets
common_org_codes = set(final_df['ORG CODE'].unique()) & set(hospital_cleaned['ORG CODE'].unique())
print(f"Number of common ORG CODEs: {len(common_org_codes)}")

# Check date ranges
print(f"final_df date range: {final_df['DATE'].min()} to {final_df['DATE'].max()}")
print(f"hospital_cleaned date range: {hospital_cleaned['DATE'].min()} to {hospital_cleaned['DATE'].max()}")

# Find overlapping date range
final_df_dates = set(final_df['DATE'].unique())
hospital_dates = set(hospital_cleaned['DATE'].unique())
common_dates = final_df_dates & hospital_dates
print(f"Number of overlapping dates: {len(common_dates)}")
if len(common_dates) > 0:
    print(f"Date overlap range: {min(common_dates)} to {max(common_dates)}")

# Memory optimization: Convert object columns to category where appropriate
print("\nOptimizing memory usage...")

# For final_df
for col in ['ORG CODE', 'CANCER TYPE', 'TRUST', 'STANDARD', 'CARE SETTING/REFERRAL ROUTE']:
    if col in final_df.columns:
        final_df[col] = final_df[col].astype('category')

# For hospital_cleaned
for col in ['ORG CODE', 'TRUST']:
    if col in hospital_cleaned.columns:
        hospital_cleaned[col] = hospital_cleaned[col].astype('category')

# Force garbage collection to free up memory (if available)
if gc_available:
    gc.collect()

# Merge the datasets
def filtered_merge(df1, df2):
    """Filter datasets to common ORG CODEs before merging"""
    print("Filtering datasets to common ORG CODEs...")

    # Filter both datasets to only include common ORG CODEs
    df1_filtered = df1[df1['ORG CODE'].isin(common_org_codes)].copy()
    df2_filtered = df2[df2['ORG CODE'].isin(common_org_codes)].copy()

    print(f"Filtered final_df shape: {df1_filtered.shape}")
    print(f"Filtered hospital_cleaned shape: {df2_filtered.shape}")

    # Perform the merge on both ORG CODE and DATE
    merged_df = pd.merge(df1_filtered, df2_filtered, on=['ORG CODE', 'DATE'], how='inner')
    return merged_df

def merge_in_chunks(df1, df2, chunk_size=50000):
    """Merge large DataFrames in chunks to avoid memory issues"""
    merged_chunks = []

    # Split df1 into chunks
    for i in range(0, len(df1), chunk_size):
        chunk = df1.iloc[i:i+chunk_size]
        print(f"Processing chunk {i//chunk_size + 1}/{(len(df1)//chunk_size) + 1}")

        # Merge chunk with hospital data on both ORG CODE and DATE
        merged_chunk = pd.merge(chunk, df2, on=['ORG CODE', 'DATE'], how='inner')
        merged_chunks.append(merged_chunk)

        # Force garbage collection after each chunk (if available)
        if gc_available:
            gc.collect()

    # Concatenate all merged chunks
    print("Concatenating chunks...")
    merged_df = pd.concat(merged_chunks, ignore_index=True)
    return merged_df

# Try the filtered merge approach first
try:
    print("\nAttempting filtered merge...")
    merged_df = filtered_merge(final_df, hospital_cleaned)
    print("✅ Filtered merge successful!")

except MemoryError:
    print("Filtered merge failed due to memory. Trying chunk-based merge...")
    try:
        merged_df = merge_in_chunks(final_df, hospital_cleaned, chunk_size=25000)
        print("✅ Chunk-based merge successful!")
    except MemoryError:
        print("Both merge strategies failed. Trying with sample...")
        sample_size = min(100000, len(final_df))
        final_df_sample = final_df.sample(n=sample_size, random_state=42)
        merged_df = pd.merge(final_df_sample, hospital_cleaned, on=['ORG CODE', 'DATE'], how='inner')
        print(f"✅ Sample merge successful with {sample_size} rows!")

print("="*60)
print("MERGE RESULTS")
print("="*60)

# Display information about the merged dataset
print(f"Merged dataset shape: {merged_df.shape}")
print(f"Memory usage: {merged_df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# Check for duplicate columns (TRUST columns from both datasets)
duplicate_cols = [col for col in merged_df.columns if col.endswith('_x') or col.endswith('_y')]
if duplicate_cols:
    print(f"Found duplicate columns: {duplicate_cols}")

    # Handle duplicate TRUST columns
    if 'TRUST_x' in merged_df.columns and 'TRUST_y' in merged_df.columns:
        merged_df = merged_df.rename(columns={
            'TRUST_x': 'CANCER_TRUST_NAME',
            'TRUST_y': 'HOSPITAL_TRUST_NAME'
        })
        print("✅ Renamed duplicate TRUST columns")

# Verify that all required features are present
print("\n" + "="*50)
print("ML FEATURE VERIFICATION")
print("="*50)

required_features = [
    'elective_overnight_admissions',     # Planned inpatient procedures
    'elective_daycase_admissions',       # Same-day procedures
    'emergency_admissions',              # Unplanned admissions
    'gp_referrals_made_total',          # Referral volume
    'first_outpatient_attendances',     # New patient volume
    'daycase_to_overnight_ratio',       # Care complexity indicator
    'emergency_to_elective_ratio',      # System pressure indicator
    'referral_conversion_rate',         # System efficiency
    'activity_diversity_index'          # Service variety measure
]

features_available = 0
print("Required ML features status:")
for feature in required_features:
    if feature in merged_df.columns:
        non_zero_count = (merged_df[feature] > 0).sum() if merged_df[feature].dtype in ['int64', 'float64'] else 'N/A'
        print(f"✓ {feature}: {non_zero_count:,} non-zero values")
        features_available += 1
    else:
        print(f"✗ {feature}: MISSING")

print(f"\nFeature availability: {features_available}/{len(required_features)} ({features_available/len(required_features)*100:.1f}%)")

if features_available >= 7:
    print("🎉 Excellent! Advanced ML analysis ready!")
elif features_available >= 5:
    print("👍 Good! Solid ML analysis possible!")
elif features_available >= 3:
    print("⚠️  Limited features - basic ML analysis possible")
else:
    print("❌ Insufficient features for meaningful ML analysis")

# Display sample of merged data
print("\nFirst 5 rows of merged dataset:")
print(merged_df.head())

print(f"\nFinal dataset columns ({len(merged_df.columns)} total):")
for i, col in enumerate(merged_df.columns.tolist(), 1):
    marker = "🎯" if col in required_features else "  "
    print(f"{marker} {i:2d}. {col}")

# Save the merged dataset
print("\n" + "="*50)
print("SAVING RESULTS")
print("="*50)

try:
    merged_df.to_excel('final_merged_complete_with_ml_features.xlsx', index=False)
    print("✅ Dataset saved as 'final_merged_complete_with_ml_features.xlsx'")
except Exception as e:
    print(f"Failed to save Excel: {e}")
    try:
        merged_df.to_csv('final_merged_complete_with_ml_features.csv', index=False)
        print("✅ Dataset saved as 'final_merged_complete_with_ml_features.csv'")
    except Exception as e:
        print(f"❌ Failed to save: {e}")

# Final memory cleanup
if gc_available:
    gc.collect()

print("\n" + "="*60)
print("MERGE COMPLETED SUCCESSFULLY! 🎉")
print("="*60)
print(f"✅ Final dataset: {merged_df.shape[0]:,} rows × {merged_df.shape[1]} columns")
print(f"✅ ML features: {features_available}/{len(required_features)} available")
print(f"✅ Date range: {merged_df['DATE'].min()} to {merged_df['DATE'].max()}")
print(f"✅ Trusts covered: {merged_df['ORG CODE'].nunique()}")
print(f"✅ Ready for ML analysis and dissertation work!")
print("="*60)
```
