```python
import pandas as pd
import re
import os

# This script is a consolidated and robust solution for loading, cleaning,
# and merging multiple Excel and CSV files into a single DataFrame.
# It addresses the issue of inconsistent header rows by automatically
# detecting the correct header position for each file.

# Define a function to standardize column names ---
def standardize_column_names(df):
    """
    Cleans and standardizes column headers in a DataFrame.
    This function converts headers to lowercase, removes special characters,
    and replaces spaces with underscores to ensure consistency across datasets.
    """
    new_columns = []
    for col in df.columns:
        # First, check if the column name is a string. If not, convert it.
        if not isinstance(col, str):
            col = str(col)

        # Convert to lowercase
        col = col.lower()
        # Replace specific characters and multiple spaces with a single underscore
        col = re.sub(r'[()\s\-\&]', '_', col).strip('_')
        # Remove any remaining non-alphanumeric characters (except underscore)
        col = re.sub(r'[^a-z0-9_]', '', col)
        # Replace multiple underscores with a single underscore
        col = re.sub(r'__+', '_', col)
        new_columns.append(col)
    df.columns = new_columns
    return df

# Define a list of files to be processed ---
# This explicit list is based on the files you mentioned and successfully downloaded.
file_list = [
    'MAR-Prov-201112-Revisions-Refresh-Jan13.xls',
    'MAR_Prov-Web-file-April-12-Rev-Jul13.xls',
    'MAR_Prov-Web-file-May-12-Rev-July-2013.xls',
    'MAR_Prov-Web-file-June-12-Rev-July-13.xls',
    'MAR_Prov-Web-file-Jul-12-Rev-July13.xls',
    'MAR_Prov-Web-file-Aug-12_37486-Rev-July13.xls',
    'MAR_Prov-Web-file-Sept-12_89770-Rev-July13.xls',
    'MAR_Prov-Web-file-Oct12_65890-Rev-July13.xls',
    'MAR_Prov-Web-file-Nov12_Rev-July13.xls',
    'MAR_Prov-Web-file-Dec12_Rev-July13.xls',
    'MAR_Prov-Web-file-Jan13_Refresh-July13.xls',
    'MAR_Prov-Web-file-Feb13_Revised-July13.xls',
    'MAR_Prov-Web-file-Mar13_Revised-July-13.xls',
    'MAR_Prov-Web-file-Apr13_68733-Jul-14-refresh.xls',
    'MAR_Prov-Web-file-May13_82947-Jul-14-refresh-2.xls',
    'MAR_Prov-Web-file-June13_93753-Jul-14-refresh.xls',
    'MAR_Prov-Web-file-July13_85428-Jul-14-refresh.xls',
    'MAR_Prov-Web-file-Aug-13-88219-Jul14-refresh.xls',
    'MAR_Prov-Web-file-Sep13_38956-Nov-15-refresh.xls',
    'MAR_Prov-Web-file-Oct13_72995-Nov-15-refresh.xls',
    'MAR_Prov-Web-file-Nov13_83795-Nov-15-refresh.xls',
    'MAR_Prov-Web-file-Dec13_77049-Nov-15-refresh.xls',
    'MAR_Prov-Web-file-Jan14_24813-Nov-15-refresh.xls',
    'MAR_Prov-Web-file-Feb14-42615-Nov-15-refresh.xls',
    'MAR_Prov-Web-file-March-14-79165-Nov-15-refresh.xls',
    'MAR_Prov-Web-file-April-14-26742-Nov-15-refresh.xls',
    'MAR_Prov-Web-file-May-14-21889-Nov-15-refresh.xls',
    'MAR_Prov-Web-file-June-14-31256-Nov-15-refresh.xls',
    'MAR_Prov-Web-file-July-14-68445-Nov-15-refresh.xls',
    'MAR_Prov-Web-file-August-14-26733-Nov-15-refresh.xls',
    'MAR_Prov-Web-file-September-14-34234-Nov-15-refresh.xls',
    'MAR_Prov-Web-file-October-14-65464-Nov-15-refresh.xls',
    'MAR_Prov-Web-file-November-14-65464-Nov-15-refresh.xls',
    'MAR_Prov-Web-file-December-14-09883-Nov-15-Refresh.xls',
    'MAR_Prov-Web-file-January-15-09883-Nov-15-Refresh.xls',
    'MAR_Prov-Web-file-February-15-092859-Nov-15-Refresh.xls',
    'MAR_Prov-Web-file-March-15-12895632-Nov-15-Refresh.xls',
    'MAR_Prov-Web-file-Apr-15-1548.xls',
    'MAR_Prov-Web-file-May-15-8456.xls',
    'MAR_Prov-Web-file-June-15-1328.xls',
    'MAR_Prov-Web-file-July-15-8846.xls',
    'MAR_Prov-Web-file-August-15-7286.xls',
    'MAR_Prov-Web-file-Sept-15-4869.xls',
    'MAR_Prov-Web-file-October-15-6412.xls',
    'MAR_Prov-Web-file-November-15-1859.xls',
    'MAR_Prov-Web-file-December-15-1213.xls',
    'MAR_Prov-Web-file-January-16-2121.xls',
    'MAR_Prov-Web-file-February-16-1213.xls',
    'MAR_Prov-Web-file-March-2016-Revised-1.xls',
    'MAR_Prov-Web-file-Apr-2016-Revised.xls',
    'MAR_Prov-Web-file-May-2016-Revised.xls',
    'MAR_Prov-Web-file-June-2016-Revised.xls',
    'MAR_Prov-Web-file-July-2016-Revised.xls',
    'MAR_Prov-Web-file-August-2016-Revised.xls',
    'MAR_Prov-Web-file-September-2016-Revised.xls',
    'MAR_Prov-Web-file-October-2016-Revised.xls',
    'MAR_Prov-Web-file-November-2016-Revised-1.xls',
    'MAR_Prov-Web-file-Decmeber-2016-Revised.xls',
    'MAR_Prov-Web-file-January-2017-Revised.xls',
    'MAR_Prov-Web-file-February-2017-Revised.xls',
    'MAR_Prov-Web-file-March-2016-Revised-2.xls',
    'MAR_Prov-Web-file-Apr-17_7uDj3K.xls',
    'MAR_Prov-Web-file-May-17_yc6SqF.xls',
    'MAR_Prov-Web-file-June-17_4KVgR3.xls',
    'MAR_Prov-Web-file-July-17_JQyh2U.xls',
    'MAR_Prov-Web-file-August-17_Eh5UKG.xls',
    'MAR_Prov-Web-file-Sept-17_Nhu2Vx.xls',
    'MAR_Prov-Web-file-Oct-17_H6fvMs.xls',
    'MAR_Prov-Web-file-Nov-17_VPxt8s.xls',
    'MAR_Prov-Web-file-Dec-17_gQAx5R.xls',
    'MAR_Prov-Web-file-Jan-18_W7cN52.xls',
    'MAR_Prov-Web-file-Feb-18_5Ub9WJ.xls',
    'MAR_Prov-Web-file-Mar-18_ktT958.xls',
    'MAR_Prov-Web-file-April-18-9tun8.xls',
    'MAR_Prov-Web-file-May-18-9tun8.xls',
    'MAR_Prov-Web-file-June-18-9tun8.xls',
    'MAR_Prov-Web-file-July-18-9tun8.xls',
    'MAR_Prov-Web-file-Aug-18-9tun8.xls',
    'MAR_Prov-Web-file-Sept-18-9tun8.xls',
    'MAR_Prov-Web-file-Oct-18-9tun8.xls',
    'MAR_Prov-Web-file-Nov-18-9tun8.xls',
    'MAR_Prov-Web-file-Dec-18-9tun8.xls',
    'MAR_Prov-Web-file-Jan-19-g3XA1.xls',
    'MAR_Prov-Web-file-Feb-19-g3XA1.xls',
    'MAR_Prov-Web-file-March-19-g3XA1.xls',
    'MAR_Prov-Web-file-Apr-19-g3XA1.xls',
    'MAR_Prov-Web-file-May-19-g3XA1.xls',
    'MAR_Prov-Web-file-Jun-19-g3XA1.xls',
    'MAR_Prov-Web-file-Jul-19-g3XA1.xls',
    'MAR_Prov-Web-file-Aug-19-g3XA1.xls',
    'MAR_Prov-Web-file-Sep-19-g3XA1.xls',
    'MAR_Prov-Web-file-Oct-19-g3XA1.xls',
    'MAR_Prov-Web-file-Nov-19-g3XA1.xls',
    'MAR_Prov-Web-file-Dec-19-g3XA1.xls',
    'MAR_Prov-Web-file-Jan-20-g3XA1.xls',
    'MAR_Prov-Web-file-Feb-20-EfWXs.xls',
    'MAR_Prov-Web-file-Mar-20-g3XA1.xls',
    'MAR_Prov-Web-file-Apr-20-g3XA1.xls',
    'MAR_Prov-Web-file-May-20-qQ0K3.xls',
    '2009_activity.xls', '2010_activity.xls',
]

# Define a known, common header to search for.
# 'Year', 'Period', and 'Org Name' are good candidates.
# We will use the standardized names for the search.
KNOWN_HEADERS = ['year', 'period', 'org name']

# Load each file and detect the correct header row ---
all_dataframes = []

print("Loading files, detecting headers, and standardizing column names...")

for file_path in file_list:
    try:
        # Read the first 20 rows of the file to search for the header
        # We need to load it into a temporary object first to detect the header.
        if file_path.endswith('.csv'):
            with open(file_path, 'rb') as f:
                temp_df = pd.read_csv(f, header=None, nrows=20, na_filter=False)
        else: # Assumes .xls or .xlsx
            with open(file_path, 'rb') as f:
                temp_df = pd.read_excel(f, header=None, nrows=20, na_filter=False)

        header_row_index = -1
        # Search for a row that contains at least one of the known headers
        for i, row in temp_df.iterrows():
            if any(isinstance(cell, str) and cell.lower().strip() in KNOWN_HEADERS for cell in row):
                header_row_index = i
                break

        if header_row_index != -1:
            # Re-read the file with the correct header index
            if file_path.endswith('.csv'):
                df = pd.read_csv(file_path, header=header_row_index)
            else:
                df = pd.read_excel(file_path, header=header_row_index)

            # Standardize the column headers
            df = standardize_column_names(df)

            # Append the cleaned DataFrame to our list
            all_dataframes.append(df)
            print(f"Successfully processed and loaded '{file_path}'. Headers found on row {header_row_index}.")
        else:
            print(f"Warning: Could not find a valid header in '{file_path}'. Skipping file.")

    except Exception as e:
        print(f"Error processing {file_path}: {e}. Skipping file.")

# Concatenate all valid DataFrames into a single one ---
if all_dataframes:
    # Concatenate the dataframes, filling any non-matching columns with NaN values.
    combined_activity = pd.concat(all_dataframes, ignore_index=True, sort=False)

    # --- Step 5: Display the first few rows and final shape to verify ---
    print("\nCombined DataFrame created successfully!")
    print("First 5 rows of the combined data:")
    print(combined_activity.head())
    print("\nShape of the combined DataFrame (rows, columns):")
    print(combined_activity.shape)

    # You can now save the combined DataFrame to a new file (e.g., a CSV)
    # combined_activity.to_csv('combined_activity_data.csv', index=False)
else:
    print("\nNo files were found or processed. The combined DataFrame could not be created.")
```

```python
# Data cleaning and preparation
# Ensure 'year' and 'period' columns are of the correct data type (string)
# and handle any potential NaNs or empty strings.
# Convert to string, strip whitespace, and handle potential NaNs
combined_activity['year'] = combined_activity['year'].astype(str).str.strip()
combined_activity['period'] = combined_activity['period'].astype(str).str.upper().str.strip()

# Define a function to determine the correct year for each row
# This function applies the financial year rule: April-December uses the start year,
# January-March uses the end year.
def get_correct_calendar_year(row):
    """
    Determines the calendar year based on the financial year and month,
    applying the rule: April-December uses the start year, January-March uses the end year.
    Returns the corrected calendar year as an integer or pd.NA for invalid inputs.
    """
    year_str = row.get('year')
    period_str = row.get('period')

    # Handle cases where 'year' or 'period' are missing, invalid strings, or 'nan'
    if pd.isna(year_str) or pd.isna(period_str) or year_str == 'nan' or period_str == 'NAN':
        return pd.NA



    # Split the financial year string (e.g., '2009-2010') into start and end years.
    try:
        if '-' in year_str:
            start_year, end_year = map(int, year_str.split('-'))
        else:
            # If 'year' is just a single year, use that year directly
            return int(year_str)
    except ValueError:
        # Handle cases where year string cannot be converted to int
        # print(f"Warning: Could not parse year string '{year_str}' for row.") # Suppress frequent warnings
        return pd.NA
    except AttributeError:
         # Handle cases where year_str might not have split method
         return pd.NA

    # Define the months that belong to the latter part of the financial year (Jan-Mar).
    winter_months = ['JANUARY', 'FEBRUARY', 'MARCH']

    # Apply the financial year rule
    if period_str in winter_months:
        return end_year
    else:
        return start_year

# Apply the function to create a new 'calendar_year' column
combined_activity['calendar_year'] = combined_activity.apply(get_correct_calendar_year, axis=1)

# Replace the original 'year' column with the new 'calendar_year'
# We convert the calendar_year to integer, coercing errors will turn NA into pandas' Int64 NA
combined_activity['year'] = combined_activity['calendar_year'].astype('Int64')

# Drop the intermediate 'calendar_year' column
combined_activity = combined_activity.drop(columns=['calendar_year'], errors='ignore')

# Display the first few rows with the modified 'year' column
display(combined_activity[['year', 'period']].head())

# Display information about the modified 'year' column
display(combined_activity['year'].info())
```

```python
# Import relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os

# Import garbage collection module for memory management
try:
    import gc
    gc_available = True
except ImportError:
    gc_available = False
    print("Warning: gc module not available, skipping garbage collection")

# Load datasets
print("Loading datasets...")
final_df = pd.read_excel('final_merged_data (3).xlsx')
hospital = pd.read_excel('hospital_activity.xlsx')

print(f"final_df shape: {final_df.shape}")
print(f"hospital shape: {hospital.shape}")

# Clean and prepare hospital dataset
print("Cleaning hospital dataset...")
hospital = hospital.sort_values(by='DATE', ascending=True).reset_index(drop=True)

# Remove rows with all NaN values
hospital = hospital.dropna(how='all')

# Define the specific activity columns we want to extract with better names
activity_columns_mapping = {
    'elective_g_a_ordinary_admissions_ffces': 'elective_overnight_admissions',
    'elective_g_a_daycase_admissions_ffces': 'elective_daycase_admissions',
    'elective_g_a_total_admissions_ffces': 'elective_total_admissions',
    'elective_g_a_planned_ordinary_admissions_ffces': 'planned_overnight_admissions',
    'elective_g_a_planned_daycase_admissions_ffces': 'planned_daycase_admissions',
    'elective_g_a_planned_total_admissions_ffces': 'planned_total_admissions',
    'elective_g_a_admissions_ffces_nhs_treatment_centres_tcs': 'treatment_centre_admissions',
    'total_non_elective_g_a_admissions_ffces': 'emergency_admissions',
    'gp_referrals_made_all_specialties': 'gp_referrals_made_total',
    'gp_referrals_seen_all_specialties': 'gp_referrals_seen_total',
    'gp_referrals_made_g_a': 'gp_referrals_made_general',
    'gp_referrals_seen_g_a': 'gp_referrals_seen_general',
    'other_referrals_made_g_a': 'other_referrals_made',
    'all_1st_outpatient_attendances_g_a': 'first_outpatient_attendances'
}

# Check which activity columns are actually available in the dataset
print("\nChecking for granular activity columns...")
available_activity_cols = []
missing_activity_cols = []

for original_col, new_col in activity_columns_mapping.items():
    if original_col in hospital.columns:
        available_activity_cols.append((original_col, new_col))
        print(f"✓ Found: {original_col} → will rename to: {new_col}")
    else:
        missing_activity_cols.append(original_col)
        print(f"✗ Missing: {original_col}")

if missing_activity_cols:
    print(f"\nMissing {len(missing_activity_cols)} activity columns. Available columns in hospital dataset:")
    print(hospital.columns.tolist())

# Rename the basic columns
hospital = hospital.rename(columns={
    'org_code': 'ORG CODE',
    'org_name': 'TRUST',
    'total_activity': 'total_hospital_activity'
})

# Rename the available granular activity columns to better names
if available_activity_cols:
    rename_dict = {original: new for original, new in available_activity_cols}
    hospital = hospital.rename(columns=rename_dict)
    print(f"\nRenamed {len(available_activity_cols)} granular activity columns with descriptive names")
else:
    print("\nWarning: No granular activity columns found! Check if column names match exactly.")

print(f"Cleaned hospital shape: {hospital.shape}")

# Display the final column structure
print(f"\nFinal hospital dataset columns:")
for col in hospital.columns:
    print(f"  - {col}")

# Check for common ORG CODEs and date ranges between datasets
common_org_codes = set(final_df['ORG CODE'].unique()) & set(hospital['ORG CODE'].unique())
print(f"\nNumber of common ORG CODEs: {len(common_org_codes)}")

# Check date ranges
print(f"final_df date range: {final_df['DATE'].min()} to {final_df['DATE'].max()}")
print(f"hospital date range: {hospital['DATE'].min()} to {hospital['DATE'].max()}")

# Find overlapping date range
final_df_dates = set(final_df['DATE'].unique())
hospital_dates = set(hospital['DATE'].unique())
common_dates = final_df_dates & hospital_dates
print(f"Number of overlapping dates: {len(common_dates)}")
if len(common_dates) > 0:
    print(f"Date overlap range: {min(common_dates)} to {max(common_dates)}")

# Memory optimization: Convert object columns to category where appropriate
print("\nOptimizing memory usage...")

# For final_df
for col in ['ORG CODE', 'CANCER TYPE', 'TRUST', 'STANDARD', 'CARE SETTING/REFERRAL ROUTE']:
    if col in final_df.columns:
        final_df[col] = final_df[col].astype('category')

# For hospital
for col in ['ORG CODE', 'TRUST']:
    if col in hospital.columns:
        hospital[col] = hospital[col].astype('category')

# Force garbage collection to free up memory (if available)
if gc_available:
    gc.collect()

# Create feature engineering functions for the granular data
def clean_and_convert_numeric(df, columns):
    """Clean and convert columns to numeric, handling mixed data types"""
    print("Cleaning and converting activity columns to numeric...")

    for col in columns:
        if col in df.columns:
            print(f"  Processing {col}...")
            # Check current data type
            original_dtype = df[col].dtype
            print(f"    Original dtype: {original_dtype}")

            # Convert to string first, then clean
            df[col] = df[col].astype(str)

            # Replace common non-numeric values
            df[col] = df[col].replace({
                'nan': '0',
                'NaN': '0',
                'NULL': '0',
                'null': '0',
                '': '0',
                '-': '0',
                'N/A': '0',
                'n/a': '0'
            })

            # Remove any non-numeric characters (keep digits, decimal points, minus signs)
            df[col] = df[col].str.replace(r'[^\d.-]', '', regex=True)

            # Convert to numeric, coercing errors to NaN
            df[col] = pd.to_numeric(df[col], errors='coerce')

            # Fill NaN with 0
            df[col] = df[col].fillna(0)

            # Ensure no negative values (shouldn't be negative activity counts)
            df[col] = df[col].abs()

            print(f"    Final dtype: {df[col].dtype}, Non-zero values: {(df[col] > 0).sum()}")

    return df

def create_derived_features(df):
    """Create derived features from granular activity data"""
    print("Creating derived features from granular activity data...")

    # First, identify all activity columns that need cleaning
    activity_cols = [col for col in df.columns if any(
        keyword in col.lower() for keyword in ['admissions', 'referrals', 'attendances']
    ) and col != 'total_hospital_activity']

    print(f"Found {len(activity_cols)} activity columns to clean: {activity_cols}")

    if activity_cols:
        # Clean and convert all activity columns to numeric
        df = clean_and_convert_numeric(df, activity_cols)

    # Now create derived features with cleaned data
    try:
        # Activity ratios
        if 'elective_overnight_admissions' in df.columns and 'elective_daycase_admissions' in df.columns:
            print("Creating daycase to overnight ratio...")
            df['daycase_to_overnight_ratio'] = (
                df['elective_daycase_admissions'] /
                (df['elective_overnight_admissions'] + 1)  # +1 to avoid division by zero
            )

        if 'emergency_admissions' in df.columns and 'elective_total_admissions' in df.columns:
            print("Creating emergency to elective ratio...")
            df['emergency_to_elective_ratio'] = (
                df['emergency_admissions'] /
                (df['elective_total_admissions'] + 1)
            )

        if 'gp_referrals_made_total' in df.columns and 'gp_referrals_seen_total' in df.columns:
            print("Creating referral conversion rate...")
            df['referral_conversion_rate'] = (
                df['gp_referrals_seen_total'] /
                (df['gp_referrals_made_total'] + 1)
            )

        # Create total activity sum from granular components (excluding ratios)
        numeric_activity_cols = [col for col in activity_cols if col in df.columns and
                               not any(word in col.lower() for word in ['ratio', 'rate', 'index', 'diversity'])]

        if len(numeric_activity_cols) > 1:
            print(f"Creating total activity from {len(numeric_activity_cols)} components...")
            df['calculated_total_activity'] = df[numeric_activity_cols].sum(axis=1)

        # Activity diversity index (Shannon entropy)
        if len(numeric_activity_cols) > 3:
            print("Creating activity diversity index...")
            activity_data = df[numeric_activity_cols].fillna(0)

            # Only calculate for rows with some activity
            row_sums = activity_data.sum(axis=1)

            # Create diversity index only for rows with activity > 0
            diversity_index = np.zeros(len(df))

            for i in range(len(df)):
                if row_sums.iloc[i] > 0:
                    # Calculate proportions
                    proportions = activity_data.iloc[i] / row_sums.iloc[i]
                    # Calculate Shannon entropy
                    entropy = -(proportions * np.log(proportions + 1e-10)).sum()
                    diversity_index[i] = entropy

            df['activity_diversity_index'] = diversity_index

        print("Feature engineering completed successfully!")

    except Exception as e:
        print(f"Error in feature engineering: {e}")
        print("Continuing without derived features...")

    return df

# Option 1: Merge in chunks to avoid memory issues
def merge_in_chunks(df1, df2, chunk_size=50000):
    """Merge large DataFrames in chunks to avoid memory issues"""
    merged_chunks = []

    # Split df1 into chunks
    for i in range(0, len(df1), chunk_size):
        chunk = df1.iloc[i:i+chunk_size]
        print(f"Processing chunk {i//chunk_size + 1}/{(len(df1)//chunk_size) + 1}")

        # Merge chunk with hospital data on both ORG CODE and DATE
        merged_chunk = pd.merge(chunk, df2, on=['ORG CODE', 'DATE'], how='inner')
        merged_chunks.append(merged_chunk)

        # Force garbage collection after each chunk (if available)
        if gc_available:
            gc.collect()

    # Concatenate all merged chunks
    print("Concatenating chunks...")
    merged_df = pd.concat(merged_chunks, ignore_index=True)
    return merged_df

# Option 2: Filter datasets first to reduce size before merging
def filtered_merge(df1, df2):
    """Filter datasets to common ORG CODEs before merging"""
    print("Filtering datasets to common ORG CODEs...")

    # Filter both datasets to only include common ORG CODEs
    df1_filtered = df1[df1['ORG CODE'].isin(common_org_codes)].copy()
    df2_filtered = df2[df2['ORG CODE'].isin(common_org_codes)].copy()

    print(f"Filtered final_df shape: {df1_filtered.shape}")
    print(f"Filtered hospital shape: {df2_filtered.shape}")

    # Create derived features before merging
    df2_filtered = create_derived_features(df2_filtered)

    # Perform the merge on both ORG CODE and DATE
    merged_df = pd.merge(df1_filtered, df2_filtered, on=['ORG CODE', 'DATE'], how='inner')
    return merged_df

# Try the filtered merge approach first (usually faster and less memory intensive)
try:
    print("\nAttempting filtered merge...")
    merged_df = filtered_merge(final_df, hospital)
    print("Filtered merge successful!")

except MemoryError:
    print("Filtered merge failed due to memory. Trying chunk-based merge...")
    try:
        # Create derived features before chunk merging
        hospital = create_derived_features(hospital)
        merged_df = merge_in_chunks(final_df, hospital, chunk_size=25000)
        print("Chunk-based merge successful!")
    except MemoryError:
        print("Both merge strategies failed. Dataset too large for current memory.")
        # Suggest sampling approach
        print("Trying with a sample of the data...")
        sample_size = min(100000, len(final_df))
        final_df_sample = final_df.sample(n=sample_size, random_state=42)
        hospital = create_derived_features(hospital)
        merged_df = pd.merge(final_df_sample, hospital, on=['ORG CODE', 'DATE'], how='inner')
        print(f"Sample merge successful with {sample_size} rows!")

# Display information about the merged dataset
print(f"\nMerged dataset shape: {merged_df.shape}")
print(f"Memory usage: {merged_df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# Check for duplicate columns (TRUST columns from both datasets)
duplicate_cols = [col for col in merged_df.columns if col.endswith('_x') or col.endswith('_y')]
if duplicate_cols:
    print(f"Found duplicate columns: {duplicate_cols}")

    # Handle duplicate TRUST columns
    if 'TRUST_x' in merged_df.columns and 'TRUST_y' in merged_df.columns:
        # Keep the TRUST column from final_df and rename the hospital one
        merged_df = merged_df.rename(columns={
            'TRUST_x': 'CANCER_TRUST_NAME',
            'TRUST_y': 'HOSPITAL_TRUST_NAME'
        })

# Display summary of the granular activity data
print("\n=== GRANULAR ACTIVITY DATA SUMMARY ===")
activity_feature_cols = [col for col in merged_df.columns if any(
    keyword in col.lower() for keyword in [
        'admissions', 'referrals', 'attendances', 'ratio', 'diversity', 'conversion'
    ]
)]

if activity_feature_cols:
    print(f"Found {len(activity_feature_cols)} activity-related features:")
    for col in sorted(activity_feature_cols):
        if col in merged_df.columns:
            non_null_count = merged_df[col].notna().sum()
            print(f"  - {col}: {non_null_count:,} non-null values")

    # Basic statistics for key activity metrics
    print(f"\nActivity Data Statistics (first 3 key metrics):")
    key_metrics = activity_feature_cols[:3]
    for metric in key_metrics:
        if metric in merged_df.columns:
            print(f"{metric}:")
            print(f"  Mean: {merged_df[metric].mean():.2f}")
            print(f"  Std:  {merged_df[metric].std():.2f}")
            print(f"  Min:  {merged_df[metric].min():.2f}")
            print(f"  Max:  {merged_df[metric].max():.2f}")
else:
    print("No granular activity features found - check if original columns existed in dataset")

# Display first few rows and basic info
print("\nFirst 5 rows of merged dataset:")
print(merged_df.head())

print(f"\nFinal column names ({len(merged_df.columns)} total):")
for i, col in enumerate(merged_df.columns.tolist(), 1):
    print(f"  {i:2d}. {col}")

print("\nDataset info:")
print(merged_df.info())

# Save the merged dataset
try:
    print("\nSaving merged dataset...")
    merged_df.to_excel('final_merged_complete_dataset_granular.xlsx', index=False)
    print("Dataset saved successfully as 'final_merged_complete_dataset_granular.xlsx'!")
except Exception as e:
    print(f"Failed to save as Excel: {e}")
    print("Trying CSV format...")
    try:
        merged_df.to_csv('final_merged_complete_dataset_granular.csv', index=False)
        print("Dataset saved as CSV!")
    except Exception as e:
        print(f"Failed to save: {e}")

# Final memory cleanup (if available)
if gc_available:
    gc.collect()
print("\nMerge process completed!")

# Summary of what we achieved
print("\n" + "="*60)
print("GRANULAR ACTIVITY EXTRACTION SUMMARY")
print("="*60)
print(f"✓ Original hospital dataset: {hospital.shape[0]:,} rows")
print(f"✓ Final merged dataset: {merged_df.shape[0]:,} rows")
print(f"✓ Total features: {merged_df.shape[1]} columns")
print(f"✓ Activity-specific features: {len(activity_feature_cols)}")
print(f"✓ Date range: {merged_df['DATE'].min()} to {merged_df['DATE'].max()}")
print(f"✓ Trusts covered: {merged_df['ORG CODE'].nunique()}")
print("✓ Ready for advanced ML analysis with granular activity patterns!")
print("="*60)
```
